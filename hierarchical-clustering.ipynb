# K-Means

```python
temp = data[['tenure', 'MonthlyCharges']]
print(temp.iloc[:, [0, 1]].values)
x = temp.iloc[:, [0, 1]].values
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
scores = []
for i in range(2, 8):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    km.fit(x)
    wcss.append(km.inertia_)

    labels = km.labels_
    silhouette_avg = silhouette_score(x, labels)
    scores.append(silhouette_avg)
    print('wcss score for n_cluster = ' + str(i) + ' is ' + str(wcss))
    print('silhoutte score for n_clusters = ' + str(i) + ' is ' + str(silhouette_avg))
```

```python
# Visualizing Elbow Method
plt.plot(range(2, 8), wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Visualizing Silhouette Score
plt.plot(range(2, 8), scores)
plt.title('Silhouette Score', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('Score')
plt.show()
```

```python
# Model Build
kmeansmodel = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 42, tol = 0.001)
y_kmeans= kmeansmodel.fit_predict(x)
```

```python
# Visualizing all the clusters

plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeansmodel.cluster_centers_[:, 0], kmeansmodel.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Tenure')
plt.ylabel('Monthly Charges')
plt.legend()
plt.show()
```

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as shc
import scipy
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import fcluster
from sklearn.preprocessing import LabelEncoder
```

```python
data
```

```python
data.nunique() # Melihat jumlah nilai unik pada setiap kolom
```

```python
# Encode data to make all columns numerical
le = LabelEncoder()
x = data.copy()

# Yes or No Columns
column = ["Dependents", "PaperlessBilling", "Churn"]
le.classes_ = np.array(["No", "Yes"])
for col in column:
  x[col] = le.transform(x[col])
```

```python
# Yes, No, No Internet Service Columns Columns
column = ["OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport"]
le.classes_ = np.array(["No internet service", "No", "Yes"])
for col in column:
  x[col] = le.transform(x[col])
```

```python
# Other Columns
column = ["Contract"]
le.classes_ = np.array(["Month-to-month", "One year", "Two year"])
for col in column:
  x[col] = le.transform(x[col])

# One Hot Encoding for Unordered Column
column = ["InternetService"]
temp = pd.get_dummies(x[column])
x.drop(column, axis = 1, inplace = True)
x = x.join(temp)
```

```python
x
```

```python
scaler = StandardScaler()
df_matrix = scaler.fit_transform(x)
```

```python
df_matrix
```

```python
length = df_matrix.shape[0]
D = scipy.zeros([length,length]) # Membuat matrix 2 dimensi dari panjang length dengan nilai set 0 semua
for i in range(length):
    for j in range(length):
        D[i,j] = scipy.spatial.distance.euclidean(df_matrix[i], df_matrix[j]) # Matrix jarak euclidean disimpan dalam variabel D
```

```python
Z1 = shc.linkage(D, 'complete')
max_clust = 4
clusters1 = fcluster(Z1, max_clust, criterion='maxclust')
clusters1
```

```python
silhouette_avg = silhouette_score(D, clusters1)
silhouette_avg
```

```python
plt.figure(figsize=(10,10))
dendro = shc.dendrogram(Z1, leaf_label_func=(lambda id: '[%s]' %id))
```

```python
Z2 = shc.linkage(D, 'single')
max_clust = 4
clusters2 = fcluster(Z2, max_clust, criterion='maxclust')
clusters2
```

```python
silhouette_avg = silhouette_score(D, clusters2)
silhouette_avg
```

```python
plt.figure(figsize=(15,15))
dendro = shc.dendrogram(Z2, leaf_label_func=(lambda id: '[%s]' %id))
```

```python
Z3 = shc.linkage(D, 'average')
max_clust = 5
clusters3 = fcluster(Z3, max_clust, criterion='maxclust')
clusters3
```

```python
silhouette_avg = silhouette_score(D, clusters3)
silhouette_avg
```

```python
plt.figure(figsize=(15,15))
dendro = shc.dendrogram(Z3, leaf_label_func=(lambda id: '[%s]' %id))
```
