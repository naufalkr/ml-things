## Preprocessing

### Split Normalized and Non Normalized Data

```python
df_normalized = df.copy()

columns_to_handle = ['Rooms', 'Distance','Postcode','Bedroom2','Bathroom','Car','Landsize','Lattitude','Longtitude','Propertycount']

mean_values = df_normalized[columns_to_handle].mean()
q1_values = df_normalized[columns_to_handle].quantile(0.25)
q3_values = df_normalized[columns_to_handle].quantile(0.75)

# Mengidentifikasi dan mengubah nilai yang melebihi q3
for column in columns_to_handle:
    mask_above_q3 = df_normalized[column] > q3_values[column]
    df_normalized.loc[mask_above_q3, column] = np.random.uniform(mean_values[column], q3_values[column], np.sum(mask_above_q3))

# Mengidentifikasi dan mengubah nilai yang kurang dari q1
for column in columns_to_handle:
    mask_below_q1 = df_normalized[column] < q1_values[column]
    df_normalized.loc[mask_below_q1, column] = np.random.uniform(q1_values[column], mean_values[column], np.sum(mask_below_q1))
```

```python
plt.figure(figsize=(20, 10))

# boxplot untuk df_normalized
plt.subplot(1, 2, 1)
df_normalized['Price'] = df_normalized['Price_log'].copy()
df_normalized = df_normalized.drop(columns=['Price_log'])
df_normalized.boxplot()
plt.title('Normalized')
plt.xticks(rotation=90)

# boxplot untuk df non normalized
plt.subplot(1, 2, 2)
df = df.drop(columns=['Price_log'])
df.boxplot()
plt.title('Non Normalized')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

### XGBoost

```python
# Non Normalized Data
xgb = XGBRegressor()
xgb.fit(X_train,Y_train)
xgb_pred = xgb.predict(X_test)

print("mean absolute error:",mean_absolute_error(Y_test, xgb_pred))
print("mean squared error:",mean_squared_error(Y_test, xgb_pred))
print("root mean squared error:",np.sqrt(mean_squared_error(Y_test, xgb_pred)))
print("r2-score:",r2_score(Y_test, xgb_pred))
print("Mean Absolute Percentage Error (MAPE):", mean_absolute_percentage_error(Y_test, xgb_pred))
```

SVM

```python
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

def objective(trial):
    try:
        C = trial.suggest_loguniform('C', 1e-2, 1e2)
        epsilon = trial.suggest_loguniform('epsilon', 1e-2, 1e0)
        kernel = trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf', 'sigmoid'])

        if kernel == 'poly':
            degree = trial.suggest_int('degree', 2, 4)
        else:
            degree = 3

        logger.info(f'Trial {trial.number}: C={C}, epsilon={epsilon}, kernel={kernel}, degree={degree}')

        svm = SVR(C=C, epsilon=epsilon, kernel=kernel, degree=degree)
        svm.fit(X_normalized_train, Y_normalized_train)

        svm_pred_normalized = svm.predict(X_normalized_test)

        mae = mean_absolute_error(Y_normalized_test, svm_pred_normalized)
        return mae

    except Exception as e:
        logger.error(f'Trial {trial.number} failed with exception: {e}')
        return float('inf')

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=10)

best_params = study.best_params
print("Best hyperparameters: ", best_params)

best_svm = SVR(**best_params)
best_svm.fit(X_normalized_train, Y_normalized_train)
best_svm_pred = best_svm.predict(X_normalized_test)

print("mean absolute error:", mean_absolute_error(Y_normalized_test, best_svm_pred))
print("mean squared error:", mean_squared_error(Y_normalized_test, best_svm_pred))
print("root mean squared error:", np.sqrt(mean_squared_error(Y_normalized_test, best_svm_pred)))
print("r2-score:", r2_score(Y_normalized_test, best_svm_pred))
print("Mean Absolute Percentage Error (MAPE):", mean_absolute_percentage_error(Y_normalized_test, best_svm_pred))
```

Random Forest

```python
def objective(trial):
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 10, 50),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'max_features': trial.suggest_categorical('max_features', ['auto', 'sqrt', 'log2'])
    }

    rfr = RandomForestRegressor(**param)

    rfr.fit(X_normalized_train, Y_normalized_train)

    rfr_pred = rfr.predict(X_normalized_test)

    mse = mean_squared_error(Y_normalized_test, rfr_pred)

    return mse

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)

best_params = study.best_params
print("Best hyperparameters: ", best_params)

best_rfr = RandomForestRegressor(**best_params)
best_rfr.fit(X_normalized_train, Y_normalized_train)
best_rfr_pred = best_rfr.predict(X_normalized_test)

print("Mean Absolute Error:", mean_absolute_error(Y_normalized_test, best_rfr_pred))
print("Mean Squared Error:", mean_squared_error(Y_normalized_test, best_rfr_pred))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(Y_normalized_test, best_rfr_pred)))
print("R2-Score:", r2_score(Y_normalized_test, best_rfr_pred))
print("Mean Absolute Percentage Error (MAPE):", mean_absolute_percentage_error(Y_normalized_test, best_rfr_pred))
```

Decision Tree

```python
def objective(trial):
    param = {
        'max_depth': trial.suggest_int('max_depth', 1, 32),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'max_features': trial.suggest_categorical('max_features', ['auto', 'sqrt', 'log2', None])
    }

    dtr = DecisionTreeRegressor(**param)

    dtr.fit(X_normalized_train, Y_normalized_train)

    dtr_pred = dtr.predict(X_normalized_test)

    mse = mean_squared_error(Y_normalized_test, dtr_pred)

    return mse

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)

# Print the best hyperparameters
best_params = study.best_params
print("Best hyperparameters: ", best_params)

best_dtr = DecisionTreeRegressor(**best_params)
best_dtr.fit(X_normalized_train, Y_normalized_train)
best_dtr_pred = best_dtr.predict(X_normalized_test)

print("Mean Absolute Error:", mean_absolute_error(Y_normalized_test, best_dtr_pred))
print("Mean Squared Error:", mean_squared_error(Y_normalized_test, best_dtr_pred))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(Y_normalized_test, best_dtr_pred)))
print("R2-Score:", r2_score(Y_normalized_test, best_dtr_pred))
print("Mean Absolute Percentage Error (MAPE):", mean_absolute_percentage_error(Y_normalized_test, best_dtr_pred))
```

KNN

```python
def objective(trial):
    n_neighbors = trial.suggest_int('n_neighbors', 1, 50)
    weights = trial.suggest_categorical('weights', ['uniform', 'distance'])
    algorithm = trial.suggest_categorical('algorithm', ['auto', 'ball_tree', 'kd_tree', 'brute'])
    leaf_size = trial.suggest_int('leaf_size', 10, 100)
    p = trial.suggest_int('p', 1, 5)

    knn = KNeighborsRegressor(
        n_neighbors=n_neighbors,
        weights=weights,
        algorithm=algorithm,
        leaf_size=leaf_size,
        p=p
    )
    knn.fit(X_normalized_train, Y_normalized_train)

    # Make predictions and evaluate the model
    knn_pred = knn.predict(X_normalized_test)
    mse = mean_squared_error(Y_normalized_test, knn_pred)

    return mse

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)  # Adjust n_trials as necessary

# Get the best hyperparameters
best_params = study.best_params
print("Best hyperparameters: ", best_params)

best_knn = KNeighborsRegressor(**best_params)
best_knn.fit(X_normalized_train, Y_normalized_train)
best_knn_pred = best_knn.predict(X_normalized_test)

print("mean absolute error:", mean_absolute_error(Y_normalized_test, best_knn_pred))
print("mean squared error:", mean_squared_error(Y_normalized_test, best_knn_pred))
print("root mean squared error:", np.sqrt(mean_squared_error(Y_normalized_test, best_knn_pred)))
print("r2-score:", r2_score(Y_normalized_test, best_knn_pred))
print("Mean Absolute Percentage Error (MAPE):", mean_absolute_percentage_error(Y_normalized_test, best_knn_pred))
```

ANN

```python
def objective(trial):
    hidden_layer_sizes = trial.suggest_categorical('hidden_layer_sizes', [(50,), (100,), (200,), (100, 50), (200, 100)])
    activation = trial.suggest_categorical('activation', ['relu', 'tanh', 'logistic'])
    solver = trial.suggest_categorical('solver', ['adam', 'sgd', 'lbfgs'])
    alpha = trial.suggest_loguniform('alpha', 1e-5, 1e-1)
    learning_rate = trial.suggest_categorical('learning_rate', ['constant', 'invscaling', 'adaptive'])

    ann = MLPRegressor(
        hidden_layer_sizes=hidden_layer_sizes,
        activation=activation,
        solver=solver,
        alpha=alpha,
        learning_rate=learning_rate,
        max_iter=2000,
        random_state=42
    )
    ann.fit(X_normalized_train, Y_normalized_train)

    ann_pred = ann.predict(X_normalized_test)
    mse = mean_squared_error(Y_normalized_test, ann_pred)

    return mse

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=10)

best_params = study.best_params
print("Best hyperparameters: ", best_params)

best_ann = MLPRegressor(**best_params, max_iter=2000, random_state=42)
best_ann.fit(X_normalized_train, Y_normalized_train)
best_ann_pred = best_ann.predict(X_normalized_test)

print("mean absolute error:", mean_absolute_error(Y_normalized_test, best_ann_pred))
print("mean squared error:", mean_squared_error(Y_normalized_test, best_ann_pred))
print("root mean squared error:", np.sqrt(mean_squared_error(Y_normalized_test, best_ann_pred)))
print("r2-score:", r2_score(Y_normalized_test, best_ann_pred))
print("Mean Absolute Percentage Error (MAPE):", mean_absolute_percentage_error(Y_normalized_test, best_ann_pred))
```
